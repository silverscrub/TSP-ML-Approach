```r
# processing time
xdata <- c(734, 1979, 6117,8246)
#NN
y1 <- c(0.0573,0.5288605928,4.329803228,9.76084075)
#LINKERN
y2 <- c(2.018, 13.29753451, 105.7316262, 202.3226726)
#2-opt
y3 <- c(3.354815435, 142.2353327, 7290.857143)
#DCN
y4 <- c(4.08,25.44, 238.25,1114)

# plot the first curve by calling plot() function
# First curve is plotted
plot(xdata, y1, type="o", col="blue", pch=1, lty=1, ylim=c(0,8000), ylab="Average Processing Time",xlab="Number of Points in TSP" )

# Add second curve to the same plot by calling points() and lines()
# Use symbol '*' for points.
points(xdata, y2, col="red", pch=2)
lines(xdata, y2, col="red",lty=2)

# Add Third curve to the same plot by calling points() and lines()
# Use symbol '+' for points.
points(xdata[-4], y3, col="dark red",pch=3)
lines(xdata[-4], y3, col="dark red", lty=3)

points(xdata, y4, col="green",pch=4)
lines(xdata, y4, col="green", lty=4)

# Adding a legend inside box at the location (2,40) in graph coordinates.
# Note that the order of plots are maintained in the vectors of attributes.
legend(1000,8000,legend=c("Nearest Neighbor","Lin-Kernighan","2-opt","D&C Neighborhood"), col=c("blue","red","black","green"),
                                   pch=c(1,2,3,4),lty=c(1,2,3,4), ncol=1)
```
